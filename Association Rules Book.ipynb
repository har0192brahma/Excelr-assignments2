## 1) Try different values of support and confidence. Observe the change in number of rules for different support,confidence values
## 2) Change the minimum length in apriori algorithm

```python
frequent_items=apriori(book,min_support=0.2,use_colnames=True)
```

```python
frequent_items
```

```python
rules= association_rules(frequent_items,metric='confidence',min_threshold=0.5)
```

```python
rules
```

```python
frequent_item_1=apriori(book,min_support=0.1,use_colnames=True)
frequent_item_1
```

```python
rules_1= association_rules(df=frequent_item_1,metric='confidence',min_threshold=0.4)
rules_1
```

```python
frequent_items_2= apriori(book,min_support=0.09,use_colnames=True)
frequent_items_2
```

```python
rules_2=association_rules(df=frequent_items_2,metric='confidence',min_threshold=0.6)
rules_2
```
